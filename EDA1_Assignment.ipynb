# Data Visualization

```python
# create histogram for numerical variables
numerical_vars = df.select_dtypes(include=['float64'])
for column in numerical_vars:
    numerical_vars.hist(figsize = (10,6) )
    plt.tight_layout()
    plt.show()
# the histogram shows that the data indicates majority of the readings fall within the normal range
```

```python
# create boxplot for numerical variables
plt.figure(figsize = (10,6))
sns.boxplot(data = df)
plt.title("Boxplot for Numerical Variables")
plt.show()
# LB : has high median range(around 130)
# (AC, FM, UC, DL, DS, DP) --> LOW OR near zero range
# ASTV has a moderate median range (around 30)
# MSTV and ALTV has very small minimal range
# MLTV has minimal median range
# Width has a moderate median range (around 50)
# Tendency, NSP has very small minimal range
```

```python
# create bar charts for catagorical variables
catagorical_vars = df.select_dtypes(include = ['object']).columns
for column in catagorical_vars:
    plt.figure(figsize = (10,6))
    df[column].value_counts().plot(kind='pie',color='blue')
    plt.ylabel('Frequency')
    plt.title('Bargraph for catagories')
    plt.show()
# There is no such catagorical varis=ables in the dataset.
```

```python
# Generate correlation heatmap
plt.figure(figsize = (14,8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True, cmap = 'Blues')
plt.title('Correlation Heatmap')
plt.show()
# Correlation Heatmap is graphical representation of correlation matrix where cell in the
# matrix is colored according to correlation b/w two vars.
# This helps in identifying patterns,relationship in the data.
```

```python
plt.figure(figsize = (15,10))
sns.pairplot(data = df)
plt.title("Pairplot")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.violinplot(data = df)
plt.xticks(rotation = 45)
plt.title("Violin Plot")
plt.show()
```
